```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("FinalAkhilMentorMindData.csv")
```

```python
df.head()
```

```python
df.set_index(df.columns[0], inplace=True)
```

```python
df.head()
```

```python
data=[]
for i in range(75):
    x=df.values[i]
#     print(~pd.isnull(x))
    x=x[~pd.isnull(x)]
    data.append(x)
data
```

```python
from markovclick import dummy
from markovclick.models import MarkovClickstream
```

```python
m = MarkovClickstream(data)
```

```python
fig = plt.figure(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='k')
sns.heatmap(m.prob_matrix, xticklabels=m.pages, yticklabels=m.pages,cmap="YlGnBu", annot=True)
```

```python
from markovclick.viz import visualise_markov_chain
graph = visualise_markov_chain(m)
```
